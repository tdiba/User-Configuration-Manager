## **Add a Setting**

```python
def add_setting(settings, setting): #Define/create a function called add_setting with two parameters. Paramters are called settings and setting
    key, value = setting #split setting into key and value
    key = key.lower() #ensures any key entered will all be lowercase regardless of the case it was entered like into the system
    value = value.lower() #makes the case of values entered all lowercase

    if key in settings:
        return f"Setting '{key}' already exists! Cannot add a new setting with this name." #if user tries to enter a key that already exists, this message will be displayed
    else:
        settings[key] = value
        return f"Setting '{key}' added with value '{value}' successfully!" #if the key doesn't exist, then it will be successfully entered into the system
```

```python
#call the function with arguments
print(add_setting(test_settings, ("volume", "high")))
```

### **Update_setting Function**

```python
# Function to update an existing setting
def update_setting(settings, setting): # settings = dictionary, setting = (key, value)
    key, value = setting               # get key and value from the input
    key = key.lower()                  # make key lowercase
    value = value.lower()              # make value lowercase

    if key in settings:                 # check if key exists
        settings[key] = value            # update the value
        return f"Setting '{key}' updated to '{value}' successfully!"
    else:
        return f"Setting '{key}' does not exist! Cannot update a non-existing setting."
```

### **Delete_setting Function**

```python
# Function to delete a setting
def delete_setting(settings, key):  # settings = dictionary, key = setting name
    key = key.lower()                # make key lowercase

    if key in settings:               # check if key exists
        del settings[key]              # remove the key-value pair
        return f"Setting '{key}' deleted successfully!"
    else:
        return "Setting not found!"
```

## **View_settings Function**

```python
# Function to view all settings
def view_settings(settings):    # settings = dictionary
    if not settings:            # checks if dictionary is empty 
        return "No settings available."
    else:
        result = "Current User Settings:\n"       # start output text

        for key, value in settings.items():
            result += f"{key.capitalize()}: {value}\n"    # format each setting

        return result                                     # returns the final formatted string
```

```python
print(view_settings(test_settings))
```

```python
##Test Updating a setting
print(update_setting(test_settings, ("theme", "dark")))
```
