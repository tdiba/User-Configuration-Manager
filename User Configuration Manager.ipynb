## **Add a Setting**

```python
def add_setting(settings, setting): #Define/create a function called add_setting with two parameters. Paramters are called settings and setting
    key, value = setting #split setting into key and value
    key = key.lower() #ensures any key entered will all be lowercase regardless of the case it was entered like into the system
    value = value.lower() #makes the case of values entered all lowercase

    if key in settings:
        return f"Setting '{key}' already exists! Cannot add a new setting with this name." #if user tries to enter a key that already exists, this message will be displayed
    else:
        settings[key] = value
        return f"Setting '{key}' added with value '{value}' successfully!" #if the key doesn't exist, then it will be successfully entered into the system
```

```python
#call the function with arguments
print(add_setting(test_settings, ("volume", "high")))
```
